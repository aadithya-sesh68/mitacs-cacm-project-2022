```python
nocrimenodedict={}
with session.begin_transaction() as tx:
    for record in tx.run("MATCH (j:Junction) WHERE NOT (j)-[:NEAREST_CRIME_JN]-(:Crime) RETURN j"):
        nocrimenodedict[record["j"]["id"]]=True
```

## For every junction that has atleast one crime id -> calculate the vulnerability score for each junction and add it to the csv file

### If the junction does not have a crime node -> for model purposes assumption is vulnerability is 0

```python
with open('junctions.csv','r') as infile, session.begin_transaction() as tx:
    dr = csv.DictReader(infile, quoting=csv.QUOTE_MINIMAL)
    for dict_row in dr:
        if int(dict_row['JunctionID']) not in nocrimenodedict: #For every junction that has atleast one crime node
            vul_score=0
            lambda_factor=0.01
            delt=1
            for record in tx.run(("MATCH (c:Crime)-[r]-(j:Junction {id: $a }) return c,r,j"),a=int(dict_row['JunctionID'])):
                vul_score+=((1-lambda_factor*delt)**(record["c"]["recency"]))
                print(record["c"]["recency"]) 
            juncdf.at[int(dict_row['JunctionID'])-1,'Vulnerability']=vul_score
tx.close()
```
